```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '../Data/Aerlingus/Aerlingus_reviews.csv'

df = pd.read_csv(file_path)
```

# Data Preprocessing

- Data was collected by web scraping from the Sky Trax website (https://www.airlinequality.com/airline-reviews/aer-lingus)

```python
def DataExploration(data):
    """
    This function explores the structure of a pandas DataFrame by displaying key information.
    
    Args:
        data (DataFrame): A pandas DataFrame to be explored.
    
    Returns:
        None: Prints the following infor about the DataFrame:
            - General info (data types, memory usage, etc.)
            - Data types of each column
            - Number of null and missing values per column
            - Shape of the DataFrame [rows, columns]
    """
    print(f"\033[1mInformation:\033[0m")
    print(f"{data.info()}\n")
    print(f"\033[1mData Types:\033[0m\n{data.dtypes}\n")
    print(f"\033[1mNull Values:\033[0m\n{data.isna().sum()}\n")
    print(f"\033[1mMissing Values:\033[0m\n{data.isnull().sum()}\n")
    print(f"\033[1mData Shape:\033[0m\n{data.shape}\n")

DataExploration(df)
```

```python
df['Overall Rating'] = df['Overall Rating'].astype(float)
df['Trip_verified'] = df['Trip_verified'].astype(bool)
df['Date Published'] = pd.to_datetime(df['Date Published'])
df['Date Flown'] = pd.to_datetime(df['Date Flown'])
df['Seat Comfort'] = df['Seat Comfort'].astype(float)
df['Cabin Staff Service'] = df['Cabin Staff Service'].astype(float)
df['Food & Beverages'] = df['Food & Beverages'].astype(float)
df['Ground Service'] = df['Ground Service'].astype(float)
df['Value For Money'] = df['Value For Money'].astype(float)
df['Inflight Entertainment'] = df['Inflight Entertainment'].astype(float)
df['Wifi & Connectivity'] = df['Wifi & Connectivity'].astype(float)
df['Aircraft'] = df['Aircraft'].astype(str
```

```python
df.info()
```

```python
missing_values = df.isnull().sum()
print("Columns with missing values:\n", missing_values[missing_values > 0])
```

```python
numerical_columns = ['Overall Rating', 'Seat Comfort', 'Cabin Staff Service', 'Food & Beverages', 
                     'Ground Service', 'Value For Money', 'Inflight Entertainment', 'Wifi & Connectivity']
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].median())
df[numerical_columns]
```

```python
# Fill missing values in categorical columns with 'Unknown'
categorical_columns = ['Aircraft', 'Type Of Traveller', 'Origin', 'Destination']
df[categorical_columns] = df[categorical_columns].fillna('Unknown')
df[categorical_columns]
```

```python
# Convert 'Date Published' to datetime format and extract Year and Month for grouping
df['Date Published'] = pd.to_datetime(df['Date Published'], errors='coerce')
df['Year'] = df['Date Published'].dt.year
df['Month'] = df['Date Published'].dt.month
```

# General Insights

```python
mean_rating = df["Overall Rating"].mean()

# Histogram for distribution of overall ratings
plt.figure(figsize=(10, 6))
plt.hist(df["Overall Rating"], bins=10, color='green', edgecolor='black', alpha=0.7)
plt.axvline(mean_rating, color='red', linestyle='dashed', linewidth=2, label=f'Mean Rating: {mean_rating:.2f}')

plt.title('Distribution of Overall Ratings')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

Distribution around the Mean: The histogram shows a concentration of ratings around the mid-to-high range, with the mean **(5.28)** positioned centrally. This suggests that most customers rate their experience as moderate to good.

However, there is a notable frequency of ratings toward the lower end, reflecting some dissatisfaction among certain passengers. This distribution suggests that while many travelers are satisfied with their experiences, there are areas where improvements can enhance overall customer satisfaction.

***

- **General Trends**: Other countries like Sweden and Vietnam also display high mean ratings (**9.00**), suggesting overall satisfaction with their experiences.
- **Need for Improvement**: Further analysis is required to identify countries with lower mean ratings, as this can provide valuable insights into areas where Aer Lingus may need to enhance its services.

This analysis highlights the importance of understanding customer satisfaction across different demographics to inform strategic improvements in service delivery.

# Temporal Analysis

```python
# Grouping by Year to calculate the mean ratings
yearly_rating = df.groupby('Year')['Overall Rating'].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(yearly_rating['Year'], yearly_rating['Overall Rating'], marker='o', color='blue', label='Mean Overall Rating')

# Avg. rating line
overall_avg_rating = yearly_rating['Overall Rating'].mean()
plt.axhline(overall_avg_rating, color='red', linestyle='dashed', linewidth=2, label=f'Overall Average Rating: {overall_avg_rating:.2f}')

plt.title('Mean Overall Ratings by Year')
plt.xlabel('Year')
plt.ylabel('Mean Overall Rating')
plt.xticks(yearly_rating['Year'], rotation=45)
plt.legend()
plt.grid()


'''------------------------------------------------------------------------------------------------------------'''


# mean ratings for each month across all years
average_monthly_rating = df.groupby('Month')['Overall Rating'].mean().reset_index()

# Convert month numbers to names
average_monthly_rating['Month'] = average_monthly_rating['Month'].apply(lambda x: pd.to_datetime(f'2022-{x}-01').strftime('%B'))


plt.figure(figsize=(12, 6))
plt.bar(average_monthly_rating['Month'], average_monthly_rating['Overall Rating'], color='green', edgecolor='black')

plt.title('Average Overall Ratings by Month')
plt.xlabel('Month')
plt.ylabel('Average Overall Rating')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

The analysis of mean overall ratings by year shows the following insights:

- **Fluctuations**: The average ratings fluctuates over the years, showing variability in customer satisfaction.
<br>

- **Trends**: A downward trend is clear from around **2014** onwards, suggesting that further investigation may be needed to understand the factors contributing to this decline.
<br>

- **Overall Average Rating**: The overall average rating across all years is approximately **4.97**, indicating a generally moderate level of customer satisfaction.

This plot highlights the importance of continuous monitoring of customer feedback to enhance service delivery and address potential issues.

<br>

The analysis of average overall ratings by month reveals the following insights:

- **Monthly Variation**: The average ratings noticebly differ across months.
<br>

- **Higher Ratings**: **January** and **February** have the highest average ratings, suggesting potential seasonal factors or events that enhance customer experiences during these months. Post holiday discounts and promotions can improve perceived value, while a more relaxed atmosphere on less crowded flights after holidays gives added comfort and attentive service from staff. Many passengers might also be in a positive mindset, travelling to see family or returning from holiday breaks, which can enhance their satisfaction. Additionally, companies including airlines often roll out new improvements at the start of the year, such as updated amenities, contributing to a heightened travel experiences.
<br>

- **Lower Ratings**: Conversely, **May**, **June** and **December** show lowest average ratings. these times of the year that might benefit from further analysis into customer dissatisfaction. These months coincide with peak travel seasons, which can lead to crowded flights, longer wait times, and potentially reduced comfort, impacting passenger satisfaction. December’s holiday rush may also strain airline and airport services, while June sees high summer travel demand, which can stretch resources. In addition, seasonal price increases or surcharges during these peak times may influence the customers perceived value, contributing to Aerlingus recieving a lower ratings.

This analysis underscores the importance of monitoring customer feedback throughout the year to identify trends and enhance service delivery.

# Traveller Demographics

```python
# Group by tyep of traveller to calculate mean ratings
traveler_type_rating = df.groupby('Type Of Traveller')['Overall Rating'].mean().reset_index()

# Sort values 
traveler_type_rating = traveler_type_rating.sort_values(by='Overall Rating', ascending=False)

plt.figure(figsize=(12, 6))
plt.barh(traveler_type_rating['Type Of Traveller'], traveler_type_rating['Overall Rating'], color='green', edgecolor='black')
plt.xlabel('Mean Overall Rating')
plt.title('Mean Overall Ratings by Type of Traveller')
plt.grid(axis='x', alpha=0.75)
plt.show()
traveler_type_rating.head() 
```

The analysis of mean overall ratings by type of traveler reveals the following insights:

- **Solo Leisure Travelers**: This group enjoys the highest average rating of **4.71**, reflecting a positive experience.
- **Business Travelers**: With an average rating of **4.13**, they are generally satisfied but indicate areas for potential enhancement.
- **Family Leisure Travelers**: Their average rating of **3.74** suggests moderate satisfaction, with opportunities for improvement.
- **Couple Leisure Travelers**: This group has the lowest average rating at **3.43**, indicating a need for targeted improvements in services for couples.

Understanding these differences can help Aer Lingus tailor its services and address the specific needs of each traveler category to enhance overall customer satisfaction.


```python
def map_aircraft_family(aircraft):
    if 'A320' in aircraft or 'A321' in aircraft:
        return 'Airbus A320 Family'
    elif 'A330' in aircraft:
        return 'Airbus A330 Family'
    elif 'Boeing' in aircraft:
        return 'Boeing'
    else:
        return 'Other'

# mapping to make a new column for aircraft families
df['Aircraft Family'] = df['Aircraft'].apply(map_aircraft_family)

# group by aircraft family to calculate mean ratings and counts
family_rating = df.groupby('Aircraft Family')['Overall Rating'].agg(['count', 'mean']).reset_index()

# Sort most common aircraft families
family_rating = family_rating.sort_values(by='count', ascending=False)

plt.figure(figsize=(12, 6))
plt.barh(family_rating['Aircraft Family'], family_rating['mean'], color='green', edgecolor='black')
plt.xlabel('Mean Overall Rating')
plt.title('Mean Overall Ratings by Aircraft Family')
plt.grid(axis='x', alpha=0.75)
plt.show()

family_rating.head()
```

The analysis of mean overall ratings by aircraft family provides the following insights:

- **Airbus A320 Family**: This family has the highest average rating of **5.80**, indicating strong customer satisfaction.
- **Boeing**: With an average rating of **5.67**, this family also reflects positive feedback, although it has a smaller sample size.
- **Airbus A330 Family**: Averaging **5.27**, this group suggests there are areas for improvement to enhance customer experiences.
- **Other**: This category has the highest number of reviews (750) but a moderate average rating of **5.19**, indicating that while it is popular, customer satisfaction may vary widely.

Grouping aircraft types in this manner allows for more straightforward comparisons and can help Aer Lingus focus on enhancing services for specific aircraft families to improve overall customer satisfaction.


# Service Analysis

```python
service_columns = [
    'Seat Comfort',
    'Cabin Staff Service',
    'Food & Beverages',
    'Inflight Entertainment',
    'Ground Service',
    'Overall Rating'
]

# Calculate the correlation matrix
correlation_matrix = df[service_columns].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation between Service Aspects and Overall Ratings')
plt.show()

correlation_matrix
```

The correlation analysis between specific service aspects and overall ratings reveals the following insights:

- **Strongest Correlations**:
  - **Cabin Staff Service**: Correlation of **0.77**, indicating that positive experiences significantly enhance overall satisfaction.
  - **Seat Comfort**: Correlation of **0.75**, suggesting comfort is a key factor in ratings.
  - **Food & Beverages**: Correlation of **0.66**, showing satisfaction with food options contributes positively.

- **Moderate Correlations**:
  - **Ground Service**: Correlation of **0.80**, underscoring its importance in enhancing passenger satisfaction.
  - **Inflight Entertainment**: Weakest correlation at **0.26**, suggesting it is less influential compared to other service factors.

This analysis highlights the need for Aer Lingus to focus on improving cabin staff service, seat comfort, and food & beverage options to enhance overall customer satisfaction effectively.


```python
# Calculate the average ratings for each service aspect
average_service_ratings = df[service_columns].mean().reset_index()
average_service_ratings.columns = ['Service Aspect', 'Average Rating']

# Get the services with the lowest average rating
lowest_rating_aspect = average_service_ratings.loc[average_service_ratings['Average Rating'].idxmin()]

lowest_rating_aspect
```

The analysis reveals that **Ground Service** has the lowest average rating at **2.32**. This indicates a considerable level of dissatisfaction among passengers regarding the ground service experience.

Improving the quality and efficiency of ground services could enhance overall customer satisfaction and positively impact ratings for Aer Lingus.

# Customer Behaviour

```python
# Group type of traveller to calculate the average overall rating and count for each type
traveller_satisfaction = df.groupby('Type Of Traveller')['Overall Rating'].agg(['mean', 'count']).reset_index()
traveller_satisfaction = traveller_satisfaction.sort_values(by='mean', ascending=False)

# Calculate the recommendation rate (percentage of "yes" responses) within each traveller type
recommendation_by_traveller = df.groupby('Type Of Traveller')['Recommended'].apply(lambda x: (x == "yes").mean() * 100).reset_index()
recommendation_by_traveller.columns = ['Type Of Traveller', 'Recommendation Rate (%)']

print(f'{traveller_satisfaction}\n')
print(f'{recommendation_by_traveller}')
```

```python
# Segment: High-Spending Business Travellers
high_spending_business = df[(df['Type Of Traveller'] == 'Business') & (df['Seat Type'] == 'Business Class')]

# Segment: Budget Leisure Travellers
budget_leisure = df[(df['Type Of Traveller'].isin(['Couple Leisure', 'Family Leisure'])) & (df['Seat Type'] == 'Economy Class')]

# Segment: Aspiring Comfort Seekers
aspiring_comfort_seekers = df[(df['Type Of Traveller'].isin(['Couple Leisure', 'Solo Leisure'])) &
                       (df['Seat Type'].isin(['Premium Economy', 'Economy Class'])) &
                       (df['Overall Rating'] < 4)]

segments_summary = {
    "High-Spending Business Travellers": {
        "Average Rating": high_spending_business['Overall Rating'].mean(),
        "Recommendation Rate (%)": (high_spending_business['Recommended'] == 'yes').mean() * 100,
        "Count": high_spending_business.shape[0]
    },
    "Budget Leisure Travellers": {
        "Average Rating": budget_leisure['Overall Rating'].mean(),
        "Recommendation Rate (%)": (budget_leisure['Recommended'] == 'yes').mean() * 100,
        "Count": budget_leisure.shape[0]
    },
    "Aspiring Comfort Seekers": {
        "Average Rating": aspiring_comfort_seekers['Overall Rating'].mean(),
        "Recommendation Rate (%)": (aspiring_comfort_seekers['Recommended'] == 'yes').mean() * 100,
        "Count": aspiring_comfort_seekers.shape[0]
    }
}

segments_summary_df = pd.DataFrame(segments_summary).T
segments_summary_df
```
