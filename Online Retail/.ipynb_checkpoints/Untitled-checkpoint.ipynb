```python
import pandas as pd

file1_path = '../Data/OnlineRetail/business.retailsales.csv' # Order sales
file2_path = '../Data/OnlineRetail/business.retailsales2.csv' # Annual sales

df1 = pd.read_csv(file1_path)
df2 = pd.read_csv(file2_path)
```

```python
df1, df2
```

```python
summary_stats_df1 = df1.describe()
summary_stats_df2 = df2.describe()
summary_stats_df1
```

```python
def DataExploration(data):
    """
    This function explores the structure of a pandas DataFrame by displaying key information.
    
    Args:
        data (DataFrame): A pandas DataFrame to be explored.
    
    Returns:
        None: Prints the following infor about the DataFrame:
            - General info (data types, memory usage, etc.)
            - Data types of each column
            - Number of null and missing values per column
            - Shape of the DataFrame [rows, columns]
    """
    print(f"\033[1mInformation:\033[0m")
    print(f"{data.info()}\n")
    print(f"\033[1mData Types:\033[0m\n{data.dtypes}\n")
    print(f"\033[1mNull Values:\033[0m\n{data.isna().sum()}\n")
    print(f"\033[1mMissing Values:\033[0m\n{data.isnull().sum()}\n")
    print(f"\033[1mData Shape:\033[0m\n{data.shape}\n")

DataExploration(df1)
```

```python
DataExploration(df2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Bar chart for total sales by product type
plt.figure(figsize=(12, 6))
sales_by_product = df1.groupby('Product Type')['Total Net Sales'].sum().sort_values()
sales_by_product.plot(kind='barh', color='blue')
plt.title('Total Net Sales by Product Type')
plt.xlabel('Total Net Sales')
plt.ylabel('Product Type')
plt.show()
```

```python
# 2. Line graph for total sales trend over time (using df2)
plt.figure(figsize=(12, 6))
df2['Date'] = pd.to_datetime(df2['Month'] + ' ' + df2['Year'].astype(str))
plt.plot(df2['Date'], df2['Total Sales'], marker='o')
plt.title('Total Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid()
plt.show()
```

```python
# 3. Box plot to identify outliers in sales data for df2
plt.figure(figsize=(12, 6))
sns.boxplot(x='Year', y='Total Sales', data=df2, palette='Set2')
plt.title('Box Plot of Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Calculate the average value per order for sales, discounts, and returns for each month in df2
df2['Average Gross Sales'] = df2['Gross Sales'] / df2['Total Orders']
df2['Average Discounts'] = df2['Discounts'] / df2['Total Orders']
df2['Average Returns'] = df2['Returns'] / df2['Total Orders']

# Select relevant columns for the output
average_values_per_order = df2[['Month', 'Year', 'Average Gross Sales', 'Average Discounts', 'Average Returns']]

import ace_tools as tools

# Display the results to the user
tools.display_dataframe_to_user("Average Values per Order", average_values_per_order)
```
