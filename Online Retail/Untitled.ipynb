```python
import matplotlib.pyplot as plt
import seaborn as sns

# bar chart for total sales by product type
plt.figure(figsize=(12, 6))
sales_by_product = df1.groupby('Product Type')['Returns'].sum().sort_values()
sales_by_product.plot(kind='barh', color='blue')
plt.title('Returns by Product Type')
plt.xlabel('Total Net Sales')
plt.ylabel('Product Type')
plt.show()

# bar chart for total sales by product type
plt.figure(figsize=(12, 6))
sales_by_product = df1.groupby('Product Type')['Total Net Sales'].sum().sort_values()
sales_by_product.plot(kind='barh', color='blue')
plt.title('Total Net Sales by Product Type')
plt.xlabel('Total Net Sales')
plt.ylabel('Product Type')
plt.show()
```

```python
df_products = df1.copy()

product_performance = df_products.groupby("Product Type").agg(
    Gross_Sales=('Gross Sales', 'sum'),
    Net_Sales=('Total Net Sales', 'sum')).reset_index()

# Calculate the Revenue Loss Percentage
product_performance['Revenue_Loss_Percentage'] = ((product_performance['Gross_Sales'] - product_performance['Net_Sales']) / product_performance['Gross_Sales']) * 100

# sorts products by Revenue Loss Percentage from worst performer to best
product_performance_sorted = product_performance.sort_values(by='Revenue_Loss_Percentage', ascending=False)

# colourmap 
colours = plt.cm.viridis(np.linspace(0, 1, len(product_performance_sorted)))

plt.figure(figsize=(12, 6))
plt.bar(product_performance_sorted['Product Type'], product_performance_sorted['Revenue_Loss_Percentage'], color=colours)

plt.xlabel('Product Type')
plt.ylabel('Revenue Loss Percentage (%)')
plt.title('Revenue Loss Percentage by Product Type (Worst to Best Performer)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

Graph illustrates the revenue loss percentages for various product types. Textiles are  the worst performer, with a revenue loss percentage of 11.11%, indicating that over 11% of its gross sales do not contribute to the business profitability due to discounts and returns. This is followed by the Easter product type at 10% and Music at 8.50%, both of which also reflect significant revenue loss. 

The data underscores a critical area for improvement, as high revenue loss percentages suggest that these products may be suffering from either excessive discounting strategies or high return rates from customer disatisfaction. 

Addressing the underlying issues associated with these products such as enhancing product quality, refining marketing strategies, or adjusting pricing could help minimise these losses and optimise overall sales performance.

```python
# Calculate the average value per order for sales, discounts, and returns for each month in df2
df2['Average Gross Sales'] = df2['Gross Sales'] / df2['Total Orders']
df2['Average Discounts'] = df2['Discounts'] / df2['Total Orders']
df2['Average Returns'] = df2['Returns'] / df2['Total Orders']

average_values_per_order = df2[['Month', 'Year', 'Average Gross Sales', 'Average Discounts', 'Average Returns']]
average_values_per_order
```

```python
from sklearn.linear_model import LinearRegression

# Preparing data for regression
X = np.array(range(len(df2))).reshape(-1, 1)  # Numeric representation of months
y = df2['Shipping'].values

# linear regression model
model = LinearRegression()
model.fit(X, y)

slope = model.coef_[0]
print("Average Monthly Increase in Shipping Costs (Linear Regression):", slope)

plt.figure(figsize=(12, 6))
sns.lineplot(x="Date", y="Shipping", data=df2, palette="bright", sort=False)
plt.plot(df2['Date'], model.predict(X), color='red', linestyle='--')  # Regression line
plt.title("Monthly Shipping Costs with Linear Trend")
plt.xlabel("Date")
plt.ylabel("Shipping Costs")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Regression analysis shows an average monthly increase in shipping costs of approximately 58.48. The regression line suggests a consistent upward trend in shipping expenses. 

This increase in shipping costs could impact profit margins. If these costs are rising, businesses might need to explore strategies to reduce shipping expenses or adjust pricing to maintain profitability.
