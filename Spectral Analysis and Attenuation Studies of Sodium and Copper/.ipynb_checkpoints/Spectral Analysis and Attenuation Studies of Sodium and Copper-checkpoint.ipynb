# Project Overview: Attenuation of Gamma Radiation through Copper Sheets

The linear attenuation coefficient, $\mu$, describes the attenuation of radiation as it travels through a material. In particular, an incident beam of flux $I_0$ incident upon a target of thickness $x$ is reduced to a flux $I$ leaving the target given by:

$$
I = I_0 e^{-\mu x}
$$

To measure the linear attenuation coefficient for 100 keV X-rays by copper, a series of attenuating sheets of copper of thickness 0.5 mm were placed in front of a 100 keV X-ray source, and the number of photons detected by an X-ray detector was measured in equal time intervals for different numbers of copper sheets.


### Objective
The primary objective of this project was to investigate the attenuation of gamma radiation as it passes through copper sheets of varying thicknesses. By measuring the counts of gamma radiation that penetrate these sheets, the project aimed to derive the linear attenuation coefficient for copper.

### Methodology
1. **Data Collection**: 
   - Data was collected using a gamma radiation detector placed behind copper sheets of varying thickness (measured in mm). The number of sheets was recorded, and the corresponding counts of detected radiation were measured.

2. **Data Analysis**:
   - The natural logarithm of the counts was calculated to linearize the relationship between the thickness of the copper sheets and the detected radiation, following the Beer-Lambert law.
   - A linear fit was performed to the data points representing the logarithm of the counts against the thickness of the copper sheets to estimate the slope, which corresponds to the linear attenuation coefficient.

3. **Error Analysis**:
   - The uncertainties associated with the counts were calculated using Poisson statistics, and these uncertainties were propagated through the analysis to ensure reliable estimates of the attenuation coefficient.

4. **Goodness of Fit**:
   - The quality of the linear fit was assessed using chi-squared statistics, providing insights into the reliability of the derived coefficients.

```python
# Load data
N_CU_sheets, counts = np.loadtxt("copper.txt", unpack=True)

# Calculate the thickness in mm
sheet_thickness_data = 0.5 * N_CU_sheets

ln_counts_data = np.log(counts) # log of the counts

counts_err = np.sqrt(counts) # Calculate the uncertainty in counts (Poisson statistics)
ln_counts_err = counts_err / counts # Calculate error of the natural log of counts

# plot the natural logarithim of the counts vs the thickness of the CU sheets
plt.scatter(sheet_thickness_data, ln_counts_data, label="Experiemntal data") 
plt.errorbar(sheet_thickness_data, ln_counts_data, ln_counts_err, fmt=".", capsize=5)

plt.xlabel("Thickness of CU Sheets (mm)", size=12) 
plt.ylabel("ln(counts)", size=12) 
plt.title("Attenuation", size=20)
plt.grid()
plt.legend() 
plt.show()
```

```python
# Defined fitted function
def fit_model(x, a, b): 
    return a * x + b

# Initial guesses for the parameters
initial_guess = [-0.39, 6]
popt, pcov = curve_fit(fit_model, sheet_thickness_data, ln_counts_data, initial_guess)
```

```python
# Extract fitted parameters
slope, intercept = popt
mu = -slope  # Linear attenuation coefficient
mu_error = np.sqrt(np.diag(pcov))[0]

print(f"Fitted parameters: a (slope) = {slope:.4f}, b (intercept) = {intercept:.4f}")
print(f"Linear Attenuation Coefficient (µ): {mu:.4f} ± {mu_error:.4f}")
```

The optimum value of the parameter $a$ from the fitted function $y = ax + b$ is also the slope. Given the linear attenuation equation:

$$
\ln(I) = -\mu x + \ln(I_0)
$$

we can rewrite it in the form of a linear equation:

$$
y = ax + b
$$

From the linear attenuation equation, we can see that the linear attenuation coefficient ($\mu$) is the slope of the line, and therefore $a = -\mu$.

Thus, the linear attenuation coefficient is calculated as follows:

$$
\mu = -a
$$

If the fitted parameter $a$ is found to be approximately $-0.39827412$, then:

$$
\mu = -(-0.39827412) = 0.39827412
$$



The uncertainty in the counts, which follow Poisson statistics, is calculated as:

$$
\sigma_{\text{counts}} = \sqrt{\text{counts}}
$$

This means that for each measured count, the standard deviation is equal to the square root of the count itself. This is a common approach when dealing with count data, as it reflects the inherent statistical variation in counting processes.

When we take the natural logarithm of the counts to linearize the data, we need to propagate the uncertainties. The error in the natural logarithm of the counts, denoted as $\sigma_{\ln(C)}$, can be computed using the formula for error propagation:

$$
\sigma_{\ln(C)} = \frac{\sigma_{\text{counts}}}{C}
$$

Thus, the error in the natural logarithm of counts is derived from the error in the counts. 

In our calculations, the error in the natural logarithm of counts is calculated as follows:

$$
\ln(\text{counts})_{\text{error}} = \frac{\sqrt{\text{counts}}}{\text{counts}} = \frac{1}{\sqrt{\text{counts}}}
$$

When fitting the linear model to the logarithmic data, these propagated errors are used as weights in the fitting process. This means the fit will account for the varying uncertainties associated with each measurement, leading to a more robust estimation of the linear attenuation coefficient $\mu$.


```python
x_fit = np.arange(0, 8.5, 0.5)
```

```python
plt.plot(x_fit, fit_model(x_fit, *popt), "r", label="Best fit")

# the *popt. * is there beacuse we are passing two parameters to the function
plt.scatter(sheet_thickness_data, ln_counts_data, label="Experiemntal data") 
plt.errorbar(sheet_thickness_data, ln_counts_data, ln_counts_err, fmt=".", capsize=5)
plt.xlabel("Thickness of CU Sheets (mm)", size=12) 
plt.ylabel("ln(counts)", size=12)
plt.grid()
plt.title("Attenuation", size=20)
plt.legend() 
plt.show()
```

### $\chi^2$ Test for Goodness of Fit

To assess the quality of the linear fit to the experimental data, we perform a $\chi^2$ test. The following steps outline the procedure:

1. **Calculate the Fitted Values**: 
   We first calculate the expected values for the natural logarithm of counts using the fitted model:

2. **Calculate Residuals**:
   The residuals are determined by subtracting the fitted values from the observed values:

3. **Calculate $\chi^2$**:
   The $\chi^2$ statistic is calculated using the formula:

   $$
   \chi^2 = \sum \left(\frac{\text{observed} - \text{expected}}{\text{error}}\right)^2
   $$

   This accounts for the uncertainty in the observations by dividing the residuals by their associated errors.

4. **Degrees of Freedom**:
   The degrees of freedom for the test are calculated as:

   $$
   \text{Degrees of Freedom} = N - p
   $$

   Where $N$ is the number of data points and $p$ is the number of parameters estimated (in this case, 2: the slope and intercept).

5. **Calculate the P-value**:
   The P-value is derived from the $\chi^2$ distribution, indicating the probability of observing a $\chi^2$ value as extreme as the calculated one:

   $$
   \text{P-value} = 1 - \text{CDF}(\chi^2, \text{degrees of freedom})
   $$


```python
# calculate the fitted values
ln_counts_fit = fit_model(sheet_thickness_data, *popt)

# calculate the residuals
residuals = ln_counts_data - ln_counts_fit

# calculate chi-squared
chi_squared = np.sum((residuals / ln_counts_err) ** 2)

# calculate degrees of freedom
degrees_of_freedom = len(sheet_thickness_data) - 2  # N - p (where p = 2)

# calculate P-value from chi-squared distribution
p_value = 1 - stats.chi2.cdf(chi_squared, degrees_of_freedom)


print(f"Chi-squared: {chi_squared:.4f}")
print(f"Degrees of Freedom: {degrees_of_freedom}")
print(f"P-value: {p_value:.4f}")
```

### Interpretation of Results

The results from the \( \chi^2 \) test provide important insights into the goodness of fit of the linear model to the experimental data:

1. **Chi-squared $\chi^2$**: The calculated value of **$\chi^2$ = 12.9961** indicates the level of deviation between the observed counts and those predicted by the fitted model. In general, lower $\chi^2$ values suggest a better fit, as they indicate that the observed data closely aligns with the expected data. 

2. **Degrees of Freedom**: The degrees of freedom were found to be **15**. This value is calculated as the number of data points minus the number of fitted parameters (in this case, 17 data points minus 2 parameters for the slope and intercept). Degrees of freedom are essential for interpreting the $\chi^2$ value, as they help contextualize how well the model describes the data.

3. **P-value**: The P-value of **0.6026** is significantly high, suggesting that there is no substantial evidence to reject the null hypothesis. This means that the observed data could reasonably be explained by the model. A P-value above 0.05 typically indicates that the fit is acceptable, meaning the linear model adequately describes the relationship between thickness and the logarithm of the counts.

### Conclusion
Overall, the high P-value and the relatively low $\chi^2$ value imply that the linear model is a good fit for the data, suggesting that the relationship between the thickness of the copper sheets and the attenuation of the X-ray photons is well represented by the fitted equation.

```python
# Exponential model for fitting
def exponential_model(x, I0, mu):
    return I0 * np.exp(-mu * x)
```

```python
# Perform curve fitting using the counts
initial_guess = [max(counts), 0.1]  # Initial guess for I0 and mu
popt, pcov = curve_fit(exponential_model, sheet_thickness_data, counts, p0=initial_guess)

# Extract parameters and their errors
I0, mu = popt
I0_error, mu_error = np.sqrt(np.diag(pcov))  # Standard errors from the covariance matrix
```

```python
# Calculate fitted values
counts_fit = exponential_model(sheet_thickness_data, *popt)

# Calculate residuals
residuals = counts - counts_fit

# Calculate chi-squared
chi_squared = np.sum((residuals / counts_err) ** 2)
degrees_of_freedom = len(sheet_thickness_data) - len(popt)  # N - p
p_value = 1 - stats.chi2.cdf(chi_squared, degrees_of_freedom)
```

```python
print(f"Fitted parameters: I0 = {I0:.4f} ± {I0_error:.4f}, mu = {mu:.4f} ± {mu_error:.4f}\n")
print(f"Chi-squared: {chi_squared:.4f}, Degrees of Freedom: {degrees_of_freedom}, P-value: {p_value:.4f}")

# Plot the original data with the fitted exponential curve
plt.figure(figsize=(10, 6))
plt.scatter(sheet_thickness_data, counts, label="Measured Counts", color='blue') 
plt.errorbar(sheet_thickness_data, counts, yerr=counts_err, fmt="o", capsize=5, color='blue')
plt.plot(sheet_thickness_data, counts_fit, 'r-', label='Fitted Exponential Curve')
plt.xlabel("Thickness of CU Sheets (mm)", size=12) 
plt.ylabel("Counts", size=12) 
plt.title("Fitted Exponential Model", size=20)
plt.grid()
plt.legend() 
plt.show()
```

# Part 2 Project Overview: Spectral Analysis of Sodium

A spectrometer was used to measure the spectrum of sodium in the region around 590 nm. The spectrometer produces a voltage output at each wavelength with an uncertainty due to noise of 0.1 V, and its spectral response results in the spectrum of a sharp line having a Gaussian profile of some width $\sigma_s$.

In this project, we will analyse the spectrum of sodium using the provided dataset from `spectrum.txt`. This dataset consists of wavelengths and corresponding spectrometer voltage outputs, which together represent the spectral response of sodium in a specific wavelength range.

### Steps We Will Follow

1. **Data Loading**: 
   We will load the data from the `spectrum.txt` file to access the wavelength and voltage values.
   
2. **Data Visualisation**: 
   Create plots of the spectral data to visualise the intensity as a function of wavelength, helping us identify peaks corresponding to the sodium spectral lines.

3. **Gaussian Fitting**: 
   We will fit a Gaussian function to the spectral data to model the intensity of the spectral lines. The Gaussian model is important because it can accurately represent the profile of spectral lines in the presence of instrumental broadening.

4. **Parameter Extraction**: 
   From the Gaussian fit, we will extract key parameters, including:
   - The central wavelength of the spectral line ($\lambda_0$).
   - The intensity of the line at its peak ($I_0$).
   - The width of the line, which is related to the instrument resolution.

5. **Calculation of Instrument Resolution**: 
   Determine the instrument resolution based on the fitted Gaussian parameters.

6. **Plotting Residuals**: 
   Visualise the residuals between the experimental data and the fitted Gaussian model to assess the quality of the fit.

7. **Chi-squared Test**: 
   Conduct a $\chi^2$ test to evaluate how well our fitted model describes the observed data, which helps in assessing the reliability of our Gaussian fit.

8. **Discussion of Results**: 
   Summarise the findings, including the measured wavelength of the sodium spectral line, its uncertainty, and the spectrometer's resolution.

### Importance of This Analysis

This spectral analysis is crucial for several reasons:
- **Understanding Atomic Spectra**: Analysing the spectral lines of sodium provides insights into the atomic structure and behaviour of sodium atoms, which is fundamental in fields such as atomic physics and chemistry.
- **Applications in Astrophysics**: Sodium lines are significant in astrophysics, particularly in the study of stellar atmospheres and the identification of elements in distant stars.
- **Instrument Calibration**: Accurate measurements of spectral lines help in calibrating spectroscopic instruments, ensuring their reliability in various scientific applications.


```python
# Load data
wavelength, voltage = np.loadtxt("spectrum.txt", unpack=True)

# Voltage uncertainty given to be 0.1 V
voltage_err = np.ones_like(voltage) * 0.1

# Plotting the voltage vs wavelength with error bars
plt.figure(figsize=(10, 6))
plt.scatter(wavelength, voltage, label="Experimental data", color='blue')
plt.errorbar(wavelength, voltage, yerr=voltage_err, fmt="o", capsize=5, color='blue')
plt.xlabel("Wavelength (m)", size=12)
plt.ylabel("Voltage (V)", size=12)
plt.title("Spectrum of Sodium", size=20)
plt.grid()
plt.legend()
plt.show()
```

```python
# gaussian fitting function
def fit_model_Gauss(x, A, mu, sig):
    return A * np.exp(-((x - mu) ** 2) / sig**2)


initial_guess = [2.1, 5.89000000e-07, 0.005e-7]
popt, pcov = curve_fit(fit_model_Gauss, wavelength, voltage, p0=initial_guess)

# Extract fitted parameters and their uncertainties
A_fit, mu_fit, sig_fit = popt
A_error, mu_error, sig_error = np.sqrt(np.diag(pcov))
```

```python
# subset of the total data
wavelength_data = wavelength[(wavelength > 5.88e-7) & (wavelength < 5.905e-7)]
voltage_data = voltage[(wavelength > 5.88e-7) & (wavelength < 5.905e-7)]

# Calculate fitted values using the optimal parameters
fitted_voltage = fit_model_Gauss(wavelength_data, *popt)

# Plot the original data and the fitted Gaussian curve
plt.figure(figsize=(10, 6))
plt.scatter(wavelength_data, voltage_data, label="Experimental data", color='blue')
plt.plot(wavelength_data, fitted_voltage, 'r-', label='Fitted Gaussian Curve')
plt.xlabel("Wavelength (m)", size=12)
plt.ylabel("Voltage (V)", size=12)
plt.title("Gaussian Fit to Sodium Spectrum", size=20)
plt.grid()
plt.legend()
plt.show()

print(f"Optimal parameters: \n  A = {A_fit:.4f} ± {A_error:.4f}\n  mu = {mu_fit:.4e} ± {mu_error:.4e}\n  sigma = {sig_fit:.4e} ± {sig_error:.4e}")
```

```python
x_fit_Gauss = np.arange(5.859e-07, 5.926e-07, 0.000001e-07)
 
plt.figure(figsize=(10, 6))
plt.plot(x_fit_Gauss, fit_model_Gauss(x_fit_Gauss, *popt), "r", label="Best fit", linewidth=2)
plt.scatter(wavelength, voltage, label="Experiemntal data") 
plt.errorbar(wavelength, voltage, voltage_err, fmt=".", capsize=5)
plt.xlabel("Wavelength (m)", size=12) 
plt.ylabel("Voltage (V)", size=12) 
plt.title("Gaussian Fit to Sodium Spectrum", size=20) 
plt.grid()
plt.legend() 
plt.show()
```

```python
fitted_voltage = fit_model_Gauss(wavelength, *popt)

# Calculate residuals
residuals = voltage - fitted_voltage

# Plotting the residuals
plt.figure(figsize=(10, 6))
plt.scatter(wavelength, residuals, color='blue', label='Residuals')
plt.axhline(0, color='red', linestyle='--', label='Zero Residual')
plt.xlabel("Wavelength (m)", size=12)
plt.ylabel("Residuals (V)", size=12)
plt.title("Residuals of Gaussian Fit", size=20)
plt.grid()
plt.legend()
plt.show()
```

```python
chi_squared = np.sum((residuals ** 2) / voltage_err**2)
degrees_of_freedom = len(wavelength) - len(popt)  # N - p
p_value = 1 - stats.chi2.cdf(chi_squared, degrees_of_freedom)

print(f"Chi-squared: {chi_squared:.4f}, Degrees of Freedom: {degrees_of_freedom}, P-value: {p_value:.7f}")
```

```python
def fit_model_Gauss_1(x, A1, mu1, sig1):
    return A1 * np.exp(-((x - mu1) ** 2) / sig1**2)

initialGuess1 = [2.1, 5.89000000e-07, 0.005e-7]
wavelength_data = wavelength[(wavelength > 5.88e-7) & (wavelength < 5.905e-7)]
voltage_data = voltage[(wavelength > 5.88e-7) & (wavelength < 5.905e-7)]



def fit_model_Gauss_2(x, A2, mu2, sig2):
    return A2 * np.exp(-((x - mu2) ** 2) / sig2**2)

initialGuess2 = [0.952, 5.89600870e-07, 3.96648324e-10]
wavelength_data2 = wavelength[(wavelength > 5.895e-7) & (wavelength < 5.905e-7)]
voltage_data2 = voltage[(wavelength > 5.895e-7) & (wavelength < 5.905e-7)]

popt1, pcov1 = curve_fit(fit_model_Gauss_1, wavelength_data, voltage_data, p0=initialGuess1)
popt2, pcov2 = curve_fit(fit_model_Gauss_2, wavelength_data2, voltage_data2, p0=initialGuess2)

xdata = np.linspace(5.859e-07, 5.926e-07, 1000)

# Plotting both Gaussian fits
plt.figure(figsize=(10, 6))
plt.plot(xdata, fit_model_Gauss_1(xdata, *popt1), "r-", label="Gaussian Fit 1")
plt.plot(xdata, fit_model_Gauss_2(xdata, *popt2), "g-", label="Gaussian Fit 2")
plt.scatter(wavelength, voltage, label="Experimental data", color='blue')
plt.errorbar(wavelength, voltage, voltage_err, fmt=".", capsize=3, color='blue')
plt.xlabel("Wavelength (m)", size=12)
plt.ylabel("Voltage (V)", size=12)
plt.title("Gaussian Fits to Sodium Spectrum", size=20)
plt.grid()
plt.legend()
plt.show()
```

```python
# Define the double Gaussian fitting function
def double_gaussian(x, A1, mu1, sig1, A2, mu2, sig2):
    g1 = A1 * np.exp(-((x - mu1) ** 2) / sig1**2)
    g2 = A2 * np.exp(-((x - mu2) ** 2) / sig2**2)
    return g1 + g2


initialGuess = [2.1, 5.89000000e-07, 0.005e-7, 1.4, 5.895000000e-07, 0.002e-7]
popt, pcov = curve_fit(double_gaussian, wavelength, voltage, p0=initialGuess)


# Plot fitted double Gaussian function
plt.figure(figsize=(10, 6))
plt.plot(xdata, double_gaussian(xdata, *popt), "r-", label="Double Gaussian Fit")
plt.scatter(wavelength, voltage, label="Experimental data", color='blue')
plt.errorbar(wavelength, voltage, voltage_err, fmt=".", capsize=5, color='blue')
plt.xlabel("Wavelength (m)", size=12)
plt.ylabel("Voltage (V)", size=12)
plt.title("Double Gaussian Fit to Sodium Spectrum", size=20)
plt.grid()
plt.legend()
plt.show()

# Extract fitted parameters
A1_fit, mu1_fit, sig1_fit, A2_fit, mu2_fit, sig2_fit = popt
A1_error, mu1_error, sig1_error, A2_error, mu2_error, sig2_error = np.sqrt(np.diag(pcov))

print(f"Fitted parameters:")
print(f"  A1 = {A1_fit:.4f} ± {A1_error:.4f}")
print(f"  mu1 = {mu1_fit:.4e} ± {mu1_error:.4e}")
print(f"  sig1 = {sig1_fit:.4e} ± {sig1_error:.4e}\n")
print(f"  A2 = {A2_fit:.4f} ± {A2_error:.4f}")
print(f"  mu2 = {mu2_fit:.4e} ± {mu2_error:.4e}")
print(f"  sig2 = {sig2_fit:.4e} ± {sig2_error:.4e}")
```

```python
# Calculate residuals for the double Gaussian fit
fitted_voltage_double = double_gaussian(wavelength, *popt)
residuals_double = voltage - fitted_voltage_double

# Plot residuals for the double Gaussian fit
plt.figure(figsize=(10, 6))
plt.scatter(wavelength, residuals_double, color='blue', label='Residuals')
plt.axhline(0, color='red', linestyle='--', label='Zero Residual')
plt.xlabel("Wavelength (m)", size=12)
plt.ylabel("Residuals (V)", size=12)
plt.title("Residuals of Double Gaussian Fit", size=20)
plt.grid()
plt.legend()
plt.show()
```

```python
# Calculate chi-squared for the double Gaussian fit
chi_squared_double = np.sum((residuals_double ** 2) / voltage_err**2)
degrees_of_freedom_double = len(voltage) - len(popt)  # N - p
p_value_double = 1 - stats.chi2.cdf(chi_squared_double, degrees_of_freedom_double)

print(f"Chi-squared: {chi_squared_double:.4f}, Degrees of Freedom: {degrees_of_freedom_double}, P-value: {p_value_double:.4f}")
```
